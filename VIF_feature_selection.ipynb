```python
import numpy as np
import pandas as pd
import pickle
from statsmodels.stats.outliers_influence import variance_inflation_factor
train = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/train.csv')
macro = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/macro.csv')
test = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/test.csv')
full = pd.merge(train, macro,  how='left', left_on=['timestamp'], right_on = ['timestamp'])
full = full.iloc[: , :100]
print(full.iloc[: , :100].shape)
full.fillna(0, inplace=True)
full = full._get_numeric_data()
full
```

```python
#only run once
remain_features = full.columns.values
with open('/content/drive/MyDrive/Colab Notebooks/remain_features', 'wb') as handle:
    pickle.dump(remain_features, handle, protocol=pickle.HIGHEST_PROTOCOL)
with open('/content/drive/MyDrive/Colab Notebooks/remain_features', 'rb') as handle:
    remain_features = pickle.load(handle)
print(remain_features)
```

```python
# restore previous session
with open('/content/drive/MyDrive/Colab Notebooks/remain_features', 'rb') as handle:
    remain_features = pickle.load(handle)
print(remain_features)
itr = len(full.columns.values) - len(remain_features) + 1
full = full[remain_features]
print('\n')

while True:
  print('iteraction:', itr)
  vif = pd.DataFrame()
  vif['VIF Factor'] = [variance_inflation_factor(full.values, i) for i in range(full.shape[1])]
  vif['features'] = full.columns
  index = np.argmax(vif['VIF Factor'].values)
  vif['features'].values[index]
  print(vif['features'].values[index])
  print(vif['VIF Factor'].values[index])
  if vif['VIF Factor'].values[index] <= 5 or itr == 2:
    print(vif)
    break
  new_features = np.delete(vif['features'].values, [index])
  print(new_features)
  with open('/content/drive/MyDrive/Colab Notebooks/remain_features', 'wb') as handle:
    pickle.dump(new_features, handle, protocol=pickle.HIGHEST_PROTOCOL)
  full = full[new_features]
  itr += 1
  print('\n')
```
