```python
import numpy as np
import pandas as pd
from statsmodels.stats.outliers_influence import variance_inflation_factor
train = pd.read_csv('train.csv')
macro = pd.read_csv('macro.csv')
test = pd.read_csv('test.csv')
full = pd.merge(train, macro,  how='left', left_on=['timestamp'], right_on = ['timestamp'])
full
```

```python
full.fillna(0, inplace=True)
full = full._get_numeric_data()
full
```

```python
ranked_features = []
```

```python
itr = 1
while True:
    print('iteraction:', itr)
    vif = pd.DataFrame()
    vif["VIF Factor"] = [variance_inflation_factor(full.values, i) for i in range(full.shape[1])]
    vif["features"] = full.columns
    index = np.argmax(vif['VIF Factor'].values)
    vif['features'].values[index]
    print(vif['features'].values[index])
    ranked_features.append(vif['features'].values[index])
    print(vif['VIF Factor'].values[index])
    if vif['VIF Factor'].values[index] <= 5 or itr == 371:
    break
    new_features = np.delete(vif['features'].values, [index])
    print(new_features)
    full = full[new_features]
    itr += 1
    print("\n")
```
